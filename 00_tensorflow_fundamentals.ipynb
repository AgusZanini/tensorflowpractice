### **Creating random tensors**

si no usamos ni seed de operacion ni global, vamos a obtener diferentes resultados en cada ejecucion de la operacion en cada ejecucion del programa entero.
si solo usamos seed global, vamos a obtener diferentes resultados en cada ejecucion de la operacion, pero si reiniciamos el programa vamos a obtener la misma secuencia de resultados

al usar seed de operacion, hay un contador que da un resultado para cada iteracion de la operacion, es decir, devolvera un resultado diferente por cada incrementeo del contador, cada ejecucion de la operacion, para resetear el contador hay que setear una seed global

ejemplo: si mezclamos poniendo un seed=42, cada vez que ejecutemos se va mezclar de forma diferente

```python
tf.random.shuffle(not_shuffled, seed=42)
```

pero si agragamos la seed global antes, entonces en cada re-run obtendriamos el mismo resultado

```python
tf.random.set_seed(42) # global level random seed
tf.random.shuffle(not_shuffled, seed=42) # operation level random seed
```

it looks like if we want our shuffled tensors to be in the same order, we've got to use the global level random seed as well as the operation level random seed

> Rule 4: "If both the global and the operation seed are set: Both seeds are used in conjunction to determine the random sequence."

**Matrix multiplications**

In machine learning, matrix multiplication is one of the most common tensor operations

there are two rules our tensor (or matrices) need to fulfill if we're going to matrix multiply them:

1. the inner dimensions must match
2. the resulting matrix has the shape of the outer dimensions

```python
# matrix multiplication in tensorflow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
tensor * tensor
```

```python
tensor1 = tf.constant([[1, 2, 5],
                      [7, 2, 1],
                      [3, 3, 3]])
tensor2 = tf.constant([[3, 5],
                       [6, 7],
                       [1, 8]])
tensor1, tensor2
```

```python
tf.matmul(tensor1, tensor2)
```

```python
# matrix multiplication with python operator "@"
tensor @ tensor
```

```python
tensor.shape
```

Generally, when performing matrix multiplication and one of the axes doesn't line up, you will transpose (rather than reshape) one of the tensors to satisfy the matrix multi;lication rules

Let's go through the following forms of aggregations:

### Find the positional maximum and minimum



```python
# create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# find the positional maximum
tf.argmax(F)
```

```python
# index on our largest value position
F[tf.argmax(F)]
```

```python
# find the max value of F
tf.reduce_max(F)
```

```python
# check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# find the positional minimum
tf.argmin(F)
```

```python
# find the minimum using the positional minimum index
F[tf.argmin(F)]
```
